```python
#import the Libraries
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Loading the data from csv file into a pandas DataFrame
df=pd.read_csv('cost_revenue_clean.csv')
```

```python
3.29e7
```

```python
x = DataFrame(df, columns=['production_budget_usd'])
y = DataFrame(df, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.title('Film cost vs GlobalRevenue')
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

ModelTraining : LinearRegression

```python
regression = LinearRegression()
regression.fit(x,y)
```

Slope Coefficent :

```python
regression.coef_
```

```python
#intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
#Adding the regression line here
plt.plot(x,regression.predict(x), color='red',linewidth=3)
plt.title('Film cost vs Global Revenue')
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
#getting  r square from regression
regression.score(x,y)
```
